# Preprocesamiento y transformación de los datos

# En estos datos que extraemos tenemos varias fechas faltantes, entre ellos los fin de semana

```python
# Debemos rellenar estos datos para no perjudicar a nuestros modelos
```

```python
from datetime import datetime
from pandas_datareader._utils import RemoteDataError
```

```python
START_DATE = '2019-01-01'
#END_DATE = str(datetime.now().strftime('%Y-%m-%d'))
END_DATE = '2019-12-31'
AMAZON = 'AMZN'

def get_data(ticker):
    try:
        stock_data = wb.DataReader(ticker, 'yahoo', START_DATE, END_DATE)
    except RemoteDataError:
        print('No dara found for {t}'.format(t=ticker))
    return stock_data
```

```python
stock=get_data(AMAZON)
print(stock.shape)
stock.describe()
```

```python
stock.head()
```

```python
days = pd.date_range(start=START_DATE, end=END_DATE)
stock = stock.reindex(days)
```

```python
# Vemos como falta infromación de los fines de semana
stock.head(10)
```

```python
# Usamos el método interpolate para rellenar los días sin información
# Esto hará una interpolación de los datos llenando los datos faltantes
stock = stock.interpolate(method='time')
```

```python
stock.isnull().sum()
```

```python
# Vimos que sigue habiendo un dato faltante, este es el 1-1-2019, usamos el metodo fillna para llenarlo
stock = stock.fillna(method='bfill')
```

```python
#Calculamos la media movil de 20 y 10 dias
rolling_data_20 = stock.Close.rolling(window=20, min_periods=1).mean()
rolling_data_10 = stock.Close.rolling(window=10, min_periods=1).mean()
plt.plot(rolling_data_20)
plt.plot(rolling_data_10)
plt.plot(stock.Close)
plt.legend(['media 20 dias', 'media 10 dias', 'price'])
plt.show()
```

# Filtro de Kalman

```python
from pykalman import KalmanFilter
from scipy import poly1d
```

```python
x = get_data(AMAZON)
x = x.Close
```

```python
kf = KalmanFilter(transition_matrices = [1],
                 observation_matrices = [1],
                 initial_state_mean = 0,
                 initial_state_covariance = 1,
                 observation_covariance = 1,
                 transition_covariance = 0.5)

# Usamos los valores observados de el precio para obtener una media movil
state_means, _ = kf.filter(x.values)
state_means = pd.Series(state_means.flatten(), index=x.index)

# Calcular la media mocil con varias ventanas hacia atrás
mean10 = x.rolling(window = 10).mean()
mean30 = x.rolling(window = 30).mean()
mean60 = x.rolling(window = 60).mean()
mean90 = x.rolling(window = 90).mean()
```

```python
state_means.shape # esto es para crear una serie temporal
```

```python
# Graficamos los datos originales y la media estimada
plt.plot(state_means)
plt.plot(x)
plt.plot(mean30)
plt.plot(mean60)
plt.plot(mean90)
plt.title('Kalman filter estimate of average')
plt.legend(['Kalman Estimate', 'X', '30-day Moving Average', '60-day Moving Average', '90-day Moving Average'])
plt.xlabel('Day')
plt.ylabel('Price');
```
